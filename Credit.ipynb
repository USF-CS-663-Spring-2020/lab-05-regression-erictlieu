```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('credit.csv')
df
```

```python
df.plot(kind='scatter', x='Income', y='Limit')
df.plot(kind='scatter', x='Rating', y='Limit')
df.plot(kind='scatter', x='Cards', y='Limit')
df.plot(kind='scatter', x='Age', y='Limit')
df.plot(kind='scatter', x='Education', y='Limit')
```

```python
sns.lmplot(x='Income', y='Limit', data=df)
sns.lmplot(x='Rating', y='Limit', data=df)
sns.lmplot(x='Cards', y='Limit', data=df)
sns.lmplot(x='Age', y='Limit', data=df)
sns.lmplot(x='Education', y='Limit', data=df)
```

```python
r1 = sns.residplot(x='Income', y='Limit', data=df)
r1.set(xlabel='Income', ylabel='Residuals')
```

```python
r2 = sns.residplot(x='Rating', y='Limit', data=df)
r2.set(xlabel='Rating', ylabel='Residuals')
```

```python
r3 = sns.residplot(x='Cards', y='Limit', data=df)
r3.set(xlabel='Cards', ylabel='Residuals')
```

```python
r4 = sns.residplot(x='Age', y='Limit', data=df)
r4.set(xlabel='Age', ylabel='Residuals')
```

```python
r5 = sns.residplot(x='Education', y='Limit', data=df)
r5.set(xlabel='Education', ylabel='Residuals')
```

```python
from sklearn.linear_model import LinearRegression

lin_reg = LinearRegression()
lin_reg.fit(df[['Income']], df.iloc[:,2])
lin_reg
print(lin_reg.coef_)
print(lin_reg.intercept_)
```

```python
lin_reg.fit(df[['Rating']], df.iloc[:,2])
lin_reg
print(lin_reg.coef_)
print(lin_reg.intercept_)
```

```python
lin_reg.fit(df[['Cards']], df.iloc[:,2])
lin_reg
print(lin_reg.coef_)
print(lin_reg.intercept_)
```

```python
lin_reg.fit(df[['Age']], df.iloc[:,2])
lin_reg
print(lin_reg.coef_)
print(lin_reg.intercept_)
```

```python
lin_reg.fit(df[['Education']], df.iloc[:,2])
lin_reg
print(lin_reg.coef_)
print(lin_reg.intercept_)
```

```python
#The predictors Income and Rating are suitable for simple linear regression
#The predictors Cards, Age, and Education are not suitable for simple linear regression
#The residual graph for Cards appear to not have homoscedasticity
#The residual graph for Age and Education appear to not have a normal distribution
```
